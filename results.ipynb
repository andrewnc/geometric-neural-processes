```python
common_precision_min = []
common_precision_mean = []
common_precision_max = []
common_precision_std = []


common_recall_min = []
common_recall_mean = []
common_recall_max = []
common_recall_std = []


random_precision_min = []
random_precision_mean = []
random_precision_max = []
random_precision_std = []


random_recall_min = []
random_recall_mean = []
random_recall_max = []
random_recall_std = []


rf_precision_min = []
rf_precision_mean = []
rf_precision_max = []
rf_precision_std = []


rf_recall_min = []
rf_recall_mean = []
rf_recall_max = []
rf_recall_std = []


gnp_precision_min = []
gnp_precision_mean = []
gnp_precision_max = []
gnp_precision_std = []


gnp_recall_min = []
gnp_recall_mean = []
gnp_recall_max = []
gnp_recall_std = []




for data_amount in data_amounts:
    common_prec_li = [x['macro avg']['precision'] for x in df[df['data_amount'] == data_amount]['common_cr']]
    common_recall_li = [x['macro avg']['recall'] for x in df[df['data_amount'] == data_amount]['common_cr']]
    common_precision_min.append(np.min(common_prec_li))
    common_precision_mean.append(np.mean(common_prec_li))    
    common_precision_max.append(np.max(common_prec_li))
    common_precision_std.append(np.var(common_prec_li))
    
    
    common_recall_min.append(np.min(common_recall_li))
    common_recall_mean.append(np.mean(common_recall_li))    
    common_recall_max.append(np.max(common_recall_li))
    common_recall_std.append(np.var(common_recall_li))
    
    
    
    random_prec_li = [x['macro avg']['precision'] for x in df[df['data_amount'] == data_amount]['random_cr']]
    random_recall_li = [x['macro avg']['recall'] for x in df[df['data_amount'] == data_amount]['random_cr']]
    random_precision_min.append(np.min(random_prec_li))
    random_precision_mean.append(np.mean(random_prec_li))    
    random_precision_max.append(np.max(random_prec_li))
    random_precision_std.append(np.var(random_prec_li))
    
    
    random_recall_min.append(np.min(random_recall_li))
    random_recall_mean.append(np.mean(random_recall_li))    
    random_recall_max.append(np.max(random_recall_li))
    random_recall_std.append(np.var(random_recall_li))
    
    
    rf_prec_li = [x['macro avg']['precision'] for x in df[df['data_amount'] == data_amount]['rf_cr']]
    rf_recall_li = [x['macro avg']['recall'] for x in df[df['data_amount'] == data_amount]['rf_cr']]
    rf_precision_min.append(np.min(rf_prec_li))
    rf_precision_mean.append(np.mean(rf_prec_li))    
    rf_precision_max.append(np.max(rf_prec_li))
    rf_precision_std.append(np.var(rf_prec_li))
    
    
    rf_recall_min.append(np.min(rf_recall_li))
    rf_recall_mean.append(np.mean(rf_recall_li))    
    rf_recall_max.append(np.max(rf_recall_li))
    rf_recall_std.append(np.var(rf_recall_li))
    
    
    
    gnp_prec_li = [x['precision'] for x in df_results[df_results['data_amount'] == data_amount]['gnp_cr']]
    gnp_recall_li = [x['recall'] for x in df_results[df_results['data_amount'] == data_amount]['gnp_cr']]
    gnp_precision_min.append(np.min(gnp_prec_li))
    gnp_precision_mean.append(np.mean(gnp_prec_li))    
    gnp_precision_max.append(np.max(gnp_prec_li))
    gnp_precision_std.append(np.var(gnp_prec_li))
    
    
    gnp_recall_min.append(np.min(gnp_recall_li))
    gnp_recall_mean.append(np.mean(gnp_recall_li))    
    gnp_recall_max.append(np.max(gnp_recall_li))
    gnp_recall_std.append(np.var(gnp_recall_li))
    
    
```

```python

df = pd.read_csv("GNP results - Sheet1.csv")

labels = ["AIDS",
"bzr_md",
"cox2_md",
"dhfr_md",
"er_md",
"mutagenicity",
"mutag",
"PTC_FM",
"PTC_FR",
"PTC_MM",
"tox21_ahr",
"Tox21_ARE",
"Tox21_AR-LBD",
"Tox21_aromatase",
"Tox21_ATAD5",
"Tox21_ER"]


x = np.array(list(range(16)))

width = .35

fig, ax = plt.subplots()

ax.plot(x, df['gnp_f1'], marker='o', markersize=3, lw=2, label="Graph Neural Process")
ax.fill_between(x, df['gnp_f1'] - df['gnp_f1_std'], df['gnp_f1'] + df['gnp_f1_std'], alpha=.25)

ax.plot(x, df['rf_f1'], marker='v',markersize=3, lw=2, label="Random Forest")
ax.fill_between(x, df['rf_f1'] - df['rf_f1_std'], df['rf_f1'] + df['rf_f1_std'], alpha=.25)

ax.plot(x, df['common_f1'], marker='D',markersize=3, lw=2,  label="Common Value")
ax.fill_between(x, df['common_f1'] - df['common_f1_std'], df['common_f1'] + df['common_f1_std'], alpha=.25)


ax.plot(x, df['random_f1'], marker='*', markersize=3, lw=2, label="Random Value")
ax.fill_between(x, df['random_f1'] - df['random_f1_std'], df['random_f1'] + df['random_f1_std'], alpha=.25)


plt.xticks(list(range(len(labels))), labels)
plt.ylabel("f1 score")
fig.autofmt_xdate()
plt.legend(ncol=2, loc="center left", bbox_to_anchor=(.048, -.45))
plt.savefig("f1_score.png", dpi=300, bbox_inches='tight')
plt.show()

fig, ax = plt.subplots()

ax.plot(x, df['gnp_rec'], marker='o',markersize=3, lw=2,  label="Graph Neural Process")
ax.fill_between(x, df['gnp_rec'] - df['gnp_prec_std'], df['gnp_rec'] + df['gnp_prec_std'], alpha=.25)

ax.plot(x, df['rf_prec'], marker='v',markersize=3, lw=2, label="Random Forest")
ax.fill_between(x, df['rf_prec'] - df['rf_prec_std'], df['rf_prec'] + df['rf_prec_std'], alpha=.25)


ax.plot(x, df['common_prec'], marker='D',markersize=3, lw=2,  label="Common Value")
ax.fill_between(x, df['common_prec'] - df['common_prec_std'], df['common_prec'] + df['common_prec_std'], alpha=.25)

ax.plot(x, df['random_prec'], marker='*',markersize=3, lw=2,  label="Random Value")
ax.fill_between(x, df['random_prec'] - df['random_prec_std'], df['random_prec'] + df['random_prec_std'], alpha=.25)


plt.ylabel("precision")
plt.xticks(list(range(len(labels))), labels)
fig.autofmt_xdate()
plt.legend(ncol=2, loc="center left", bbox_to_anchor=(.048, -.45))
plt.savefig("precision.png", dpi=300, bbox_inches='tight')
plt.show()

fig, ax = plt.subplots()

ax.plot(x, df['gnp_recall'], marker='o',markersize=3, lw=2,  label="Graph Neural Process")
ax.fill_between(x, df['gnp_recall'] - df['gnp_recall_std'], df['gnp_recall'] + df['gnp_recall_std'], alpha=.25)

ax.plot(x, df['rf_recall'], marker='v',markersize=3, lw=2, label="Random Forest")
ax.fill_between(x, df['rf_recall'] - df['rf_recall_std'], df['rf_recall'] + df['rf_recall_std'], alpha=.25)

ax.plot(x, df['common_recall'], marker='D',markersize=3, lw=2,  label="Common Value")
ax.fill_between(x, df['common_recall'] - df['common_recall_std'], df['common_recall'] + df['common_recall_std'], alpha=.25)

ax.plot(x, df['random_recall'], marker='*',markersize=3, lw=2,  label="Random Value")
ax.fill_between(x, df['random_recall'] - df['random_recall_std'], df['random_recall'] + df['random_recall_std'], alpha=.25)

plt.ylabel("recall")
plt.xticks(list(range(len(labels))), labels)
fig.autofmt_xdate()
plt.legend(ncol=2, loc="center left", bbox_to_anchor=(.048, -.45))
plt.savefig("recall.png", dpi=300, bbox_inches='tight')
plt.show()
```
